```python
# importing the all major libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
# Usage of inline plotting
%matplotlib inline
```

```python
dataFrame = pd.read_csv('911.csv')
```

```python
# Check head for data frame
dataFrame.head()
```

```python
# Checking the info for data frame
dataFrame.info()
```

```python
dataFrame.describe()
```

```python
# top 5 zip code who call to 911 more number of time

dataFrame['zip'].value_counts().head(5)
```

```python
# top 6 town-ship which do call frequently to 911

dataFrame['twp'].value_counts().head(6)
```

```python
# print the unique reason to call 911

dataFrame['title'].unique()
```

```python
# print the number of unique reason to call 911
dataFrame['title'].nunique()
```

```python
# print top 5 reason to call in 911 Use 0 or 1 in split

dataFrame['SpecificReason'] = dataFrame['title'].apply(lambda title : title.split(':')[1])
```

```python
dataFrame.head()
```

```python
dataFrame['SpecificReason'].value_counts().head(10)
```

```python
dataFrame['BroadReason'] = dataFrame['title'].apply(lambda title : title.split(':')[0])
```

```python
dataFrame['BroadReason'].value_counts().head(5)
```

```python
dataFrame.head(10)
```

## Ploting Count Plot 

```python
# Creating Count Plot fro Broad Reason
sns.countplot(x = 'BroadReason', data = dataFrame)
```

## Date-Time

```python
# convert time Stamp from object to actual Time Stamp
dataFrame['timeStamp'] = pd.to_datetime(dataFrame['timeStamp'])
```

```python
dataFrame['timeStamp']
```

```python
dataFrame.info()
```

```python
type(dataFrame['timeStamp'].iloc[0])
```

```python
# Creating 3 columns for hrs, months and days of week

dataFrame['Hour'] = dataFrame['timeStamp'].apply(lambda time:time.hour)
```

```python
dataFrame.head()
```

```python
dataFrame['Month'] = dataFrame['timeStamp'].apply(lambda time: time.month)
```

```python
dataFrame.head(10)
```

```python
dataFrame['Day'] = dataFrame['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
dataFrame.head()
```

```python
# printing the unique values of day, month and hour

dataFrame['Day'].nunique()
```

```python
dataFrame['Hour'].nunique()
```

```python
dataFrame['Month'].nunique()
```

## Ploting the graph for Month 

```python
sns.countplot(x = 'Month', data = dataFrame , hue = 'BroadReason')
# printing the legend outside the graph we use anchor tag 
plt.legend(bbox_to_anchor =(1.30, 1.0))
```

```python
# Use GroupBy function prinitng the data in perfect order by data month hour

byMonth = dataFrame.groupby('Month').count()
```
